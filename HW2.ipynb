```python
import os # подключаем модуль os
os.chdir("C:/Users/User/Desktop/poems") #выбираем директорию
print(os.path) #печатаем директорию
ans = 0
for root, dirs, files in os.walk('C:/Users/User/Desktop/poems'):
    for name in files:
        fullname = os.path.join(root, name)
        print(fullname) #печатаем названия файлов
        f = open(fullname, 'r', encoding='utf-8') #открываем файл
        poema= f.read()
        with open('poems.txt', 'a', encoding='utf-8') as b:
            b.write(poema)
        f.close()
        b.close()
    ans += len(files) #заодно считаем количество файлов в каждой папке
print(ans)
```

```python
text = "Удивительная кошка \
Несчастная кошка порезала лапу- \
Сидит, и ни шагу не может ступить. \
Скорей, чтобы вылечить кошкину лапу \
Воздушные шарики надо купить! \
И сразу столпился народ на дороге- \
Шумит, и кричит, и на кошку глядит. \
А кошка отчасти идет по дороге, \
Отчасти по воздуху плавно летит!"
print("Длина текста в символах - ", len(text)) #считаем и печатаем длину текста
wordds = [word.strip("!?.,-") for word in text.split()] #очищаем текст от пунктуации
print(wordds)
strroka1 = ' '.join(wordds) # объединяем слова в строку через пробел
print(strroka1)
wordds1=([word.lower() if word.istitle() else word for word in wordds]) #проверяем, написано ли слово с большой буквы, если да- меняем на строчную
print(wordds1)
print("Длина текста в словах - ",len(wordds1)) #печатаем длину текста в словах
s = 0
for ij in range (len(wordds1)):            #проходим циклом по всем словам
       if wordds.count(wordds[ij]) == 1:   #если слово уникальное, увеличиваем счетчик уникальных слов
            s = s+  1
print("Количество уникальных слов -",s)
print(wordds[-10:]) #печатаем последние 10 слов
strroka2 = '\n'.join(wordds) #собираем все в строку через перенос строки
print(strroka2)
print(strroka2[19:30]) #распечатываем с 20 по 30 символ этой строки (включительно)
mass = []
for l in range(len(wordds) - 1):  #создаем кортеж биграмм
    mass.append(wordds[l] + ' ' + wordds[l + 1])
mass = tuple(mass)
for n in range (len(wordds) - 1):
    print(mass[n] + "-" + str(mass.count(mass[n]))) #печатаем биграммы по порядку и количество их употребления
```
